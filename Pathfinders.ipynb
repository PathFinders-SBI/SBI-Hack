```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.utils.class_weight import compute_class_weight
from tensorflow.keras import layers, models
import tensorflow as tf
```

```python
X = df.drop(columns=['TARGET'])
y = df['TARGET']

# STEP 2: Train-Test Split (Stratified for imbalance)
X_train, X_val, y_train, y_val = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_val_scaled = scaler.transform(X_val)
```

```python
# STEP 4: Compute Class Weights
class_weights_array = compute_class_weight(class_weight='balanced', classes=np.unique(y_train), y=y_train)
class_weight_dict = dict(enumerate(class_weights_array))
print("✅ Class Weights:", class_weight_dict)

# STEP 5: Build the Neural Network Model
def build_nn_model():
    model = models.Sequential([
    layers.Dense(128, activation='relu', input_shape=(X_train_scaled.shape[1],)),
    layers.Dropout(0.3),
    layers.Dense(64, activation='relu'),
    layers.Dropout(0.2),
    layers.Dense(32, activation='relu'),
    layers.Dense(1, activation='sigmoid')
])
# STEP 4: Compute Class Weights
class_weights_array = compute_class_weight(class_weight='balanced', classes=np.unique(y_train), y=y_train)
class_weight_dict = dict(enumerate(class_weights_array))
print("✅ Class Weights:", class_weight_dict)

# STEP 5: Build the Neural Network Model
def build_nn_model():
    model = models.Sequential([
    layers.Dense(128, activation='relu', input_shape=(X_train_scaled.shape[1],)),
    layers.Dropout(0.3),
    layers.Dense(64, activation='relu'),
    layers.Dropout(0.2),
    layers.Dense(32, activation='relu'),
    layers.Dense(1, activation='sigmoid')
])

# STEP 6: Compile the Model
    model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy', tf.keras.metrics.AUC(name='auc')]
)
    return model

# STEP 7: Train the Model with Validation
nn = build_nn_model()
history = nn.fit(
    X_train_scaled, y_train,
    epochs=50,
    batch_size=512,
    validation_data=(X_val_scaled, y_val),
    class_weight=class_weight_dict,
    verbose=2
)

# STEP 8: Save the Model
nn.save("defaulter_model.h5")
print("📦 Model saved as defaulter_model.h5")
```

```python
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import precision_recall_curve, precision_score, recall_score, roc_auc_score

# 1) Prepare storage for out-of-fold predictions
n_train = X_train_scaled.shape[0]
oof_probs = np.zeros((n_train, 3))  # col0: NN, col1: LGBM, col2: XGB

kf = StratifiedKFold(n_splits=2, shuffle=True, random_state=42)
for fold, (train_idx, val_idx) in enumerate(kf.split(X_train_scaled, y_train)):
    # a) Train fresh copies on the train-fold
    nn_fold = build_nn_model()
    nn_fold.fit(
        X_train_scaled[train_idx],
        y_train.iloc[train_idx],
        epochs=50,
        batch_size=512,
        class_weight=class_weight_dict,
        verbose=0
    )

    lgbm_fold = lgb.LGBMClassifier(
        n_estimators=200,
        class_weight='balanced',
        random_state=42
    )
    lgbm_fold.fit(
        X_train_scaled[train_idx],
        y_train.iloc[train_idx]
    )

    xgb_fold = xgb.XGBClassifier(
        n_estimators=200,
        use_label_encoder=False,
        eval_metric='logloss',
        random_state=42
    )
    xgb_fold.fit(
        X_train_scaled[train_idx],
        y_train.iloc[train_idx]
    )

    # b) Predict on the fold’s validation slice
    oof_probs[val_idx, 0] = nn_fold.predict(X_train_scaled[val_idx]).ravel()
    oof_probs[val_idx, 1] = lgbm_fold.predict_proba(X_train_scaled[val_idx])[:, 1]
    oof_probs[val_idx, 2] = xgb_fold.predict_proba(X_train_scaled[val_idx])[:, 1]

# 2) Train meta-learner on those OOF predictions
meta = LogisticRegression(class_weight='balanced', max_iter=1000)
meta.fit(oof_probs, y_train.values)

# 3) Build meta-features for your held-out validation set
val_probs_nn   = nn.predict(X_val_scaled).ravel()
val_probs_lgbm = lgbm.predict_proba(X_val_scaled)[:, 1]
val_probs_xgb = xgb_model.predict_proba(X_val_scaled)[:, 1]
meta_features  = np.vstack([val_probs_nn, val_probs_lgbm, val_probs_xgb]).T

# 4) Meta-model’s final probabilities
final_probs = meta.predict_proba(meta_features)[:, 1]

# 5) Threshold tuning
precision, recall, thresholds = precision_recall_curve(y_val, final_probs)
target_recall = 0.6457
ix = np.argmin(np.abs(recall - target_recall))
opt_thresh = thresholds[ix]

# 6) Evaluate at that threshold
y_pred = (final_probs >= opt_thresh).astype(int)
print(f"Blended Precision: {precision_score(y_val, y_pred):.3f}")
print(f"Blended Recall   : {recall_score   (y_val, y_pred):.3f}")
print(f"Blended ROC AUC  : {roc_auc_score(y_val, final_probs):.3f}")
```

```python
import pandas as pd
import numpy as np
import re

# Load test data
test_df = pd.read_csv("/content/HACKATHON_PREDICTION_DATA.CSV")

# --- Store and drop UNIQUE_ID ---
unique_ids = test_df['UNIQUE_ID']
test_df = test_df.drop(columns=['UNIQUE_ID'])


# Convert 'x yrs y mon' format to months
def convert_to_months(x):
    if pd.isna(x):
        return np.nan
    match = re.match(r'(\d+)yrs\s+(\d+)mon', x)
    return int(match.group(1)) * 12 + int(match.group(2)) if match else np.nan

test_df['AVERAGE_ACCT_AGE1'] = test_df['AVERAGE_ACCT_AGE1'].apply(convert_to_months)
test_df['CREDIT_HISTORY_LENGTH1'] = test_df['CREDIT_HISTORY_LENGTH1'].apply(convert_to_months)
test_df['AVERAGE_ACCT_AGE1'].fillna(test_df['AVERAGE_ACCT_AGE1'].median(), inplace=True)
test_df['CREDIT_HISTORY_LENGTH1'].fillna(test_df['CREDIT_HISTORY_LENGTH1'].median(), inplace=True)

# Map income band
income_band_map = {
    'A': 1, 'B': 2, 'C': 3, 'D': 4, 'E': 5, 'F': 6, 'G': 7, 'H': 8,
    'EX01': 9, 'EX02': 10, 'EX04': 11, 'EX05': 12
}
test_df['INCOME_BAND1'] = test_df['INCOME_BAND1'].map(income_band_map)
test_df['INCOME_BAND1'].fillna(test_df['INCOME_BAND1'].median(), inplace=True)

# One-hot encode categorical features
test_df = pd.get_dummies(test_df, columns=['AGREG_GROUP', 'PRODUCT_TYPE', 'TIME_PERIOD'], drop_first=True)

# Binary flags
binary_flags = ['SI_FLG', 'LOCKER_HLDR_IND', 'UID_FLG', 'KYC_FLG', 'INB_FLG', 'EKYC_FLG']
test_df = test_df[~test_df['KYC_FLG'].isin(['1', '2'])]
for col in binary_flags:
    test_df[col] = test_df[col].map({'Y': 1, 'N': 0})

# Monthly features fill with 0
monthly_cols = [col for col in test_df.columns if any(key in col for key in [
    'MNTHSCR', 'MNTHSDR', 'AVGMTD', 'AVGQTD', 'AVGYTD', 'OUTSTANGBAL'])]
test_df[monthly_cols] = test_df[monthly_cols].fillna(0)

# Bureau columns fill with median
bureau_cols = [col for col in test_df.columns if any(key in col for key in [
    'CRIFF_', 'PRI_', 'PRIMARY_INSTAL_AMT', 'TOTAL_CRIFF1', 'DEC_CRIFFCHNG1'])]
test_df[bureau_cols] = test_df[bureau_cols].fillna(test_df[bureau_cols].median())

# --- Add UNIQUE_ID back ---
test_df['UNIQUE_ID'] = unique_ids

# Display the first few rows to verify
display(test_df.head())
```

```python
import joblib
from sklearn.impute import SimpleImputer

# Load the saved blended model pipeline
blend_pipe = joblib.load('defaulter_blend.pkl')

# Extract the components
nn_model = blend_pipe['nn_model']
lgbm_model = blend_pipe['lgbm_model']
xgb_model = blend_pipe['xgb_model']
meta_model = blend_pipe['meta_model']
scaler = blend_pipe['scaler']
opt_thresh = blend_pipe['threshold']

# Align columns with the training data
X_train_columns = [col for col in df.columns if col != 'TARGET']
test_df = test_df.reindex(columns=X_train_columns, fill_value=0)

# Scale the preprocessed test data
X_test_scaled = scaler.transform(test_df)

# Predict probabilities with each base model
test_probs_nn = nn_model.predict(X_test_scaled).ravel()
test_probs_lgbm = lgbm_model.predict_proba(X_test_scaled)[:, 1]
test_probs_xgb = xgb_model.predict_proba(X_test_scaled)[:, 1]

# Create meta-features for the test set
test_meta_features = np.vstack([test_probs_nn, test_probs_lgbm, test_probs_xgb]).T

# Impute any missing values in the meta-features
imputer = SimpleImputer(strategy='mean')
test_meta_features = imputer.fit_transform(test_meta_features)

# Predict with the meta-model
final_test_probs = meta_model.predict_proba(test_meta_features)[:, 1]

# Apply the optimal threshold to get final predictions
final_predictions = (final_test_probs >= opt_thresh).astype(int)

# Create the submission DataFrame
submission_df = pd.DataFrame({
    'UNIQUE_ID': test_df['UNIQUE_ID'],
    'TARGET': final_predictions
})

# Display the first few rows of the submission file
display(submission_df.head())

# Save the submission file
submission_df.to_csv('submission2.csv', index=False)

print("✅ Submission file 'submission2.csv' created successfully.")
```
